```python
plt.figure(figsize=(10, 6))

# Assuming you have a DataFrame named df
sns.lineplot(x='Brand', y='Avg Daily Distance (km)', marker='o', color='red', data=csv)

plt.xlabel('Brand')
plt.ylabel('Avg Daily Distance (km)')
plt.title('Average Daily Distance by Brand')
plt.show()
```

# Price Distribution Across Different Brands (INR)

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))  # Initialize figure
sns.swarmplot(x=csv['Owner Type'].astype(str), y=csv['Resale Price (INR)'])  # Convert Owner Type to string
plt.xticks(rotation=45)  # Rotate x-axis labels
plt.grid()  # Add grid
plt.show()  # Show plot
```

```python
df=csv.groupby('Owner Type')['Resale Price (INR)'].count()
df
```
